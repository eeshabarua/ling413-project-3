```python
import pandas as pd
```

```python
# Load metadata csv
filepath = "../gutenberg/metadata/metadata.csv"
df = pd.read_csv(filepath)
df
```

```python
import pandas as pd
# Extracting the string within the curly braces
df['subjects'] = df['subjects'].str.extract(r"{(.*?)}")

# Splitting the extracted string by commas and expanding into separate columns
split_columns = df['subjects'].str.split(', ', expand=True)

# Stack the columns to have each tag in its own row
stacked_tags = split_columns.stack().reset_index(level=1, drop=True).rename('tag')

# Concatenate the stacked tags with the original DataFrame
df = df.join(stacked_tags)
```

# Grouping All Texts by Tag 

```python
# Grouping rows by common tags
grouped = df.groupby('tag').agg(list)
```

```python
grouped
```

```python
df.to_csv("grouped_df/grouped_by_subject.csv")
```

# Grouping Poetry Texts by tag 

```python
# Fill NaN values with empty strings
df['tag'] = df['tag'].fillna('')

# Filtering rows where the tag contains "Poetry"
filtered_df = df[df['tag'].str.contains('Poetry', case=False)]

# Display the filtered DataFrame
# print(filtered_df)
```

```python
# Grouping poetry rows by common tags
grouped_poetry = filtered_df.groupby('tag').agg(list)
```
