```python
import pandas as pd
import os
```

```python
genre_name = "'Poetry'"

in_filepath = f"poems/{genre_name}/"
out_filepath = f'df_cleaned/{genre_name}/'
```

```python
def list_csv_files(directory):
    text_files = []
    for file in os.listdir(directory):
        if file.endswith(".csv"):
            text_files.append(file)
    return text_files
```

```python
csv_files = list_csv_files(in_filepath)
print(csv_files[0])
```

```python
for csv_file in csv_files:
    df = pd.read_csv(in_filepath+csv_file)
    df.drop(columns=['Unnamed: 0'], inplace=True)

    # let's drop all rows with empty poem title
    df.dropna(inplace=True)
    
    # Filter out rows where Stanza has 10 or less words
    df = df[df['Stanza'].apply(lambda x: len(x.split()) >= 10)]
    
    # Save 
```

```python
df.to_csv(out_filepath+csv_files[0])
```
