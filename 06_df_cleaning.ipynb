```python
import pandas as pd
import os
```

```python
genre_name = "'American poetry'"

in_filepath = f"poems/{genre_name}/"
out_filepath = f'df_cleaned/{genre_name}/'
os.makedirs(f'{out_filepath}/{genre_name}', exist_ok=True)

delimiter = 'line'
```

```python
def list_csv_files(directory):
    text_files = []
    for file in os.listdir(directory):
        if file.endswith(f"_{delimiter}.csv"):
            text_files.append(file)
    return text_files
```

```python
csv_files = list_csv_files(in_filepath)
print(csv_files[0])
```

# 1 Trimming default columns

```python
def drop_specific_rows(df):
    # Find indices of rows where 'Poem Title' is "INTRODUCTION." or "NOTES."
    indices_to_drop = df[(df['Poem Title'] == "INTRODUCTION.") | (df['Poem Title'] == "NOTES.") | (df['Poem Title'] == "FOOTNOTES:")].index 

    # Drop rows using the indices
    df.drop(indices_to_drop, inplace=True)
```

```python
for csv_file in csv_files:
    df = pd.read_csv(in_filepath+csv_file)
    df
    df.drop(columns=['Unnamed: 0'], inplace=True)

    # let's drop all rows with empty poem title
    df.dropna(inplace=True)
    
    # drop all INTRO and NOTES rows
    drop_specific_rows(df)
    
    # Filter out rows where Stanza has 10 or less words
    df = df[df[delimiter].apply(lambda x: len(x.split()) >= 10)]
    
    # Save 
```

```python
# Assuming df is your DataFrame
def print_values_ending_in_period(df):
    for column in df.columns:
        for value in df[column]:
            if isinstance(value, str) and value.endswith('.'):
                print(value)

# Call the function with your DataFrame
print_values_ending_in_period(df)
```

# 2 Data Cleaning 

```python
import re

def clean(text):
    # Define a regular expression pattern to match punctuation and special characters
    pattern = r'[^\w\s]'  # Matches any character that is not a word character or whitespace

    # Use the sub() function from the re module to replace matched characters with an empty string
    clean_text = re.sub(pattern, '', text)
    
    # Convert the cleaned text to lowercase
    clean_text_lower = clean_text.lower()
    
    # Replace underscores with spaces
    clean_text = clean_text.replace('_', ' ')

    return clean_text_lower
```

```python
df[delimiter] = df[delimiter].apply(clean)
```
