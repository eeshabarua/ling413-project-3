```python
import requests
import pandas as pd
import os
```

```python
filepath = 'raw_text'
classes = ["'Poetry'", "'English poetry'", "\"Children's poetry\"", "'American poetry'", "'English poetry -- 19th century'", "'Epic poetry", "'American poetry -- 19th century'", "'American poetry -- 20th century'", "'English poetry -- 20th century'", "1914-1918 -- Poetry'"]

genre_name = "'American poetry -- 20th century'"

# Load the PG ID corpus 
df = pd.read_csv(f"grouped_pgid_dfs/en_tag/{genre_name}.csv", index_col=0)
```

```python
# df.drop(columns='Unnamed: 0', inplace=True)
df
```

```python
def download_gutenberg_text(book_id):
    url = f'https://www.gutenberg.org/cache/epub/{book_id}/pg{book_id}.txt'
    response = requests.get(url)
    
    os.makedirs(f'{filepath}/{genre_name}', exist_ok=True)
    
    if response.status_code == 200:
        with open(f'{filepath}/{genre_name}/{book_id}.txt', 'wb') as file:
            file.write(response.content)
        print(f"File {book_id}.txt downloaded successfully.")
    else:
        print(f"Failed to download {book_id}.txt. Status code: {response.status_code}")
```

```python
# Example usage:
book_id = df.iloc[0]['id'][2:]  # Example book ID
download_gutenberg_text(book_id)
```

```python
for book in df['id']:
    book_id = book[2:]
    download_gutenberg_text(book_id)
```
