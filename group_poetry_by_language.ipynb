```python
import pandas as pd
```

```python
# Load metadata csv
filepath = "grouped_df/poetry_grouped_by_subject.csv"
df = pd.read_csv(filepath)
df.drop(columns='Unnamed: 0', inplace=True)
df
```

```python
print(set(df['language']))
```

```python
import re
for lang in set(df['language']):
    lang_str = re.sub(r"[\[\]']", '', lang)
    print(lang_str)
    lang_df = df[df['language'] == lang]
    lang_df.to_csv(f"grouped_df/lang/{lang_str}.csv")
```
