```python
import pandas as pd
```

```python
# Load metadata csv
filepath = "grouped_df/lang/en.csv"
df = pd.read_csv(filepath)
df.drop(columns='Unnamed: 0', inplace=True)
df
```

```python
# Print distribution of tags and their frequencies
tag_distribution = df['tag'].value_counts()
print(tag_distribution[:11])
```

```python
# Selecting for the top 10 most frequent classes of poetry 
classes = ["'Poetry'", "'English poetry'", "\"Children's poetry\"", "'American poetry'", "'English poetry -- 19th century'", "'Epic poetry", "'American poetry -- 19th century'", "'American poetry -- 20th century'", "'English poetry -- 20th century'", "1914-1918 -- Poetry'"]
```

```python
for cl in classes:
    print(len(df[df['tag'] == cl]))
    tag_df = df[df['tag'] == cl]
    tag_df.to_csv(f"grouped_df/en_tag/{cl}.csv")
```
